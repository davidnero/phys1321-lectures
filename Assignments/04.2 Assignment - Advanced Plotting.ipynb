# Plotting Electric Potential

The electric potential $V$ of a point charge $q$ measured a radial distance $r$ away from it is given by
$$V=k\frac{q}{r}\,,$$
where $V$ has units of volts and $k=8.99\times10^{9}$ V$\cdot$m$\cdot$C$^{-1}$ is the electrostatic constant. (We'll drop the units from here on, but assume that they're in volts, coulombs, and meters.)

Convince yourself that for a point charge $q_1$ at $(x_1, y_1)$, the electric potential at all other locations in the $xy$-plane is:
$$V(x,y)=k\frac{q_1}{\sqrt{(x-x_1)^2+(y-y_1)^2}}\,.$$

For two point charges, the electric potential is
$$V(x,y)=k\left(\frac{q_1}{\sqrt{(x-x_1)^2+(y-y_1)^2}} + \frac{q_2}{\sqrt{(x-x_2)^2+(y-y_2)^2}}\right)\,.$$
